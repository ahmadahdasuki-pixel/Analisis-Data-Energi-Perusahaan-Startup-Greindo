### **Perbandingan Kapasitas Energi Pada Masing-Masing Wilayah**

```python
#Membandingkan kapasitas terpasang antar wilayah
import matplotlib.pyplot as plt

VARIABEL = df.groupby('wilayah')['kapasitas_terpasang_mw'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(VARIABEL.index, VARIABEL.values, color='skyblue')

plt.xlabel('Wilayah')
plt.ylabel('Total Kapasitas (MEGA WATT)')
plt.title('Top 10 Wilayah berdasarkan Kapasitas')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Berdasarkan visualisasi tersebut, terlihat perbandingan kapasitas energi pada masing-masing wilayah. Secara umum, persebaran kapasitas energi masih belum merata. Beberapa wilayah memiliki kapasitas yang jauh lebih besar dibandingkan wilayah lainnya.

Ketimpangan ini dapat dipengaruhi oleh beberapa faktor, salah satunya adalah perbedaan luas wilayah. Wilayah dengan cakupan geografis yang lebih luas cenderung membutuhkan kapasitas energi yang lebih besar untuk memenuhi kebutuhan penduduk serta aktivitas industri di dalamnya. Selain itu, faktor lain seperti jumlah penduduk, tingkat urbanisasi, dan perkembangan infrastruktur juga dapat memengaruhi besarnya kapasitas energi terpasang.

Selain itu, ditemukan adanya data wilayah yang duplikat contohnya wilayah "Jawa" namun ada juga data lain yang konteks nya sama seperti wilayah "java" dan "JAVA". Keberadaan data duplikat ini dapat menyebabkan distorsi dalam analisis, seperti perhitungan kapasitas yang menjadi tidak akurat (overcounting). Hal ini berpotensi menghasilkan interpretasi yang keliru terhadap kondisi sebenarnya.

Duplikasi data tersebut kemungkinan terjadi akibat inkonsistensi dalam proses input data, perbedaan penulisan nama wilayah, atau penggabungan data dari berbagai sumber yang belum melalui proses pembersihan (data cleaning) secara optimal.

Oleh karena itu, diperlukan tahap preprocessing data yang lebih teliti, seperti normalisasi penamaan wilayah dan penghapusan data duplikat, agar hasil analisis yang diperoleh menjadi lebih valid dan dapat diandalkan.

## **Outliers Values**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Berdasarkan hasil pengecekan outliers value, dapat disimpulkan bahwa secara umum dataset memiliki jumlah outlier yang relatif kecil pada sebagian besar kolom, namun terdapat beberapa kolom yang perlu diperhatikan.

## **Outliers Values**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'laju_efesiensi': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('laju_efesiensi', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Berdasarkan hasil pengecekan outliers value, dapat disimpulkan bahwa secara umum dataset memiliki jumlah outlier yang relatif kecil pada sebagian besar kolom, namun terdapat beberapa kolom yang perlu diperhatikan.

Kolom seperti tahun, produksi_energi_gwh, emisi_karbon_terpangkas, dan skor_investasi tidak memiliki outlier (0%), yang menunjukkan bahwa distribusi data pada kolom-kolom tersebut cukup stabil dan berada dalam rentang yang wajar. Hal ini menandakan bahwa data pada kolom tersebut sudah konsisten dan tidak memerlukan penanganan khusus terkait outlier.

Namun, terdapat beberapa kolom yang memiliki outlier meskipun dalam proporsi kecil, yaitu kapasitas_terpasang_mw (0.38%), populasi_terlayani (0.50%), dan terutama laju_efisiensi (0.97%) yang memiliki persentase outlier paling tinggi di antara yang lain.

Outliers pada kolom kapasitas_terpasang_mw (0.38%) tidak perlu ditangani secara khusus karena nilai-nilai ekstrem yang muncul masih dapat dianggap wajar dalam konteks data energi.
```python
columns_to_impute = ["populasi_terlayani"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```
